```python
import pandas as pd
import os
PATH = 'data/clean'
for path, dirs, files in os.walk(PATH):
    for file in files:
        if file.endswith('.csv'):
            f = os.path.join(path, file)
            df = pd.read_csv(f)
            df = df.drop(columns=['Unnamed: 0'])
            df.to_csv(f, index=False, sep='|')
```

```python
PATH = 'data/clean'
for file in os.listdir(os.path.join(os.path.dirname(__file__), "data", "raw")):
    if file.endswith(".csv"):
        df = pd.read_csv(os.path.join(os.path.dirname(__file__), "data", "raw", file))
        df.columns = df.columns.str.lower()
        df.to_csv(os.path.join(os.path.dirname(__file__), "data", "raw", file), index=False, sep='|')
```
